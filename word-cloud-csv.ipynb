```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from PIL import Image
from wordcloud import WordCloud, ImageColorGenerator
from tabulate import tabulate  
```

```python
dataset = pd.read_csv(r'Reviews.csv\Reviews.csv', encoding='latin1')
```

```python
text2 = " ".join(dataset['Summary'].astype(str).tolist())
background_image = np.array(Image.open('355-3558165_shocked-clipart.png'))
plt.imshow(background_image)
```

```python
word_cloud2 = WordCloud(background_color='white', mask=background_image, width=2048, height=1080).generate(text2)
img_colors = ImageColorGenerator(background_image)
word_cloud2.recolor(color_func=img_colors)
word_cloud2.to_file('patrik.png')
```

```python
plt.imshow(word_cloud2, interpolation='bilinear')
plt.axis("off")
plt.show()
```
